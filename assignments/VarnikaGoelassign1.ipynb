```python
import numpy as np
```

```python
import random
```

```python
#generating random points for first cluster
firstx=[]
for i in range (100):
    firstx.append([np.random.uniform(2,4)])
firsty=[]
for i in range (100):
    firsty.append([np.random.uniform(2,4)])

#generating random points for second cluster
thirdx=[]
for i in range(100):
    thirdx.append([np.random.uniform(0,2)])
thirdy=[]
for i in range(100):
    thirdy.append([np.random.uniform(0,2)])
```

```python
import matplotlib.pyplot as plt
```

```python
#calculating mean for the clusters formed
x1=np.mean(firstx)
y1=np.mean(firsty)
x2=np.mean(thirdx)
y2=np.mean(thirdy)
```

```python
#plotting clusters and the mean obtained
plt.scatter(firstx,firsty)
plt.scatter(thirdx,thirdy)
plt.scatter(x1,y1,color="r")
plt.scatter(x2,y2,color="g")
```

```python
#finding euclidena distance
def abc(x,y):
    dist1=np.sqrt(np.square(x1-x) + np.square(y1-y))
    dist2=np.sqrt(np.square(x2-x) + np.square(y2-y))
    return dist1,dist2
x,y=3,3
dist1,dist2=abc(x,y)
```

```python
#plotting the clusters, mean and the euclidena distance
plt.scatter(firstx,firsty)
plt.scatter(thirdx,thirdy)
plt.scatter(x1,y1,color="r")
plt.scatter(x2,y2,color="g")
plt.scatter(x,y,color='k')
```

```python
if dist1<dist2:
    print("[",x,",",y,"]","belongs to cluster 1")
else:
    print("[",x,",",y,"]","belongs to cluster 2")
```
